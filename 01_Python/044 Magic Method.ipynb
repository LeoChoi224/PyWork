# `__repr__`, `__str__` 차이


```python
# __repr__(),  __str__()
# 공통점 : 객체를 문자열 리턴
# 차이점
#    `__repr__` : 표현값  (representative form)
#    `__str__` : str(), print() 등에서 문자열(str) 변환시 호출됨, 
#                 오버라이딩 안되어 있으면 __repr__ 의 값을 리턴한다
    
# https://shoark7.github.io/programming/python/difference-between-__repr__-vs-__str__
```

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
        
s1 = Student('김수진', 3)
s1        
```

```python
print(s1)
```

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def __repr__(self):
        return f"이름 {self.name}, 학년 {self.grade}"
        
s1 = Student('김수진', 3)
s1   # __repr__() 값
```

```python
print(s1)  # __str__() 값?   __str__ 이 정의 안되어 있으면 __repr__ 값이 리턴
```

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def __repr__(self):
        return f"이름 {self.name}, 학년 {self.grade}"

    def __str__(self):
        return f"[{self.name}:{self.grade}]"
        
s1 = Student('김수진', 3)
s1   # __repr__() 값
```

```python
print(s1)  # __str__() 값
```

```python
str(s1)  # __str__() 값
```

```python
f'이 것은{s1} 입니다.' # __str__()
```

```python
'이것은 %s 입니다' % (s1)
```

```python
# 연습 : 2차원 행렬 정의
# magic method 활용
"""
mt1 = Matrix2x2([10, 20], [30, 40])
mt2 = Matrix2x2([1, 2], [3, 4])

print(mt1) 실행 결과
    [10, 20]
    [30, 40] 

print(mt2) 실행 결과
    [1, 2]
    [3, 4]
    
print(mt1 + mt2) 실행결과
    [11, 22]
    [33, 44]
    
print(mt1 + mt2 + mt2 + mt2) 실행결과
    [13, 26]
    [39, 52]
"""
None
```

```python
class Matrix2x2:
    def __init__(self, row1, row2):
        self.row1 = row1
        self.row2 = row2
    
    def __repr__(self):
        return str(self.row1) + '\n' + str(self.row2) + '\n')
    
    def __add__(self, other: Matrix2x2):
        list1 = []
        list2 = []
        for i in range(len(self.row1)):
            list1.append(self.row1[i] + other.row1[i])
            list2.append(self.row2[i] + other.row2[i])
        return Matrix2x2(list1, list2)
    
    def __mul__(self, other: Matrix2x2):
        list1 = []
        list2 = []
        for i in range(len(self.row1)):
            list1.append(self.row1[i] * other.row1[i])
            list2.append(self.row2[i] * other.row2[i])
        return Matrix2x2(list1, list2)
```

```python
mt1 = Matrix2x2([10, 20], [30, 40])
mt2 = Matrix2x2([1, 2], [3, 4])

mt1.__add__(mt2)
print(mt1 + mt2)

mt1.__add__()
print(mt1)

mt1.__add__(mt2) + mt2
print(mt1 + mt2 + mt2


print(mt2)
print(mt1 + mt2)
print(mt1 + mt2 + mt2 + mt2)
print(mt1 * mt2)
print(mt1 + mt2 + mt2 * mt2)
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f'Student[name={self.name}, age={self.age}]'

    def __lt__(self, other): # "<" 연산자 정의. 참/거짓 리턴
        # age 오름차순
        # return self.age < other.age
        # 내림차순
        # return self.age > other.age
        # name 오름차순
        return self.name < other.name

s1 = Student('이브', 5)
s2 = Student('우동', 2)
s3 = Student('절미', 4)

stu_list = [s1, s2, s3]
print('정렬전', stu_list)

stu_list.sort()  
print('정렬후', stu_list)


        
        
```

# 객체 동일성 `__eq__`

```python
# 파이썬에서 “객체의 동일성(같음)”을 정의하려면 __eq__ 메서드를 구현한다. 
# __eq__ 두 객체를 ==로 비교할 때 호출되는 메서드
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Alice", 25)
p2 = Person("Alice", 25)
p3 = Person("Bob", 30)

print(p1 == p2) # 동등성
print(p1 == p3)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __eq__(self, other):   # 참/거짓 리턴
        return self.name == other.name and self.age == other.age

p1 = Person("Alice", 25)
p2 = Person("Alice", 25)
p3 = Person("Bob", 30)

print(p1 == p2) # 동등성  True
print(p1 == p3)
```

# dataclass 사용

```python
from dataclasses import dataclass

@dataclass
class Person:
    name: str
    age: int

# 이렇게 하면:

# __eq__
# __init__
# __repr__

# 자동 생성됨
```

```python
p1 = Person("Alice", 25)
p2 = Person("Alice", 25)
p3 = Person("Bob", 30)

print(p1 == p2)  # True
print(p1 == p3)  # False
```
