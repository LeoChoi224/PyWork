# 단축키들

```python
# command mode : 셀 변경 추가 이동..
# edit mode : 코드 작성, Markdown 작성

# edit mode -> command mode [ESC]
#    SHIFT + ENTER : 코드 실행

# command mode -> edit mode [클릭 or ENTER]
#    A :  위에 셀 추가
#    B :  아래에 셀 추가
#    X :  셀 삭제
#    Z : UNDO
#    Y :  --> Code 셀로 변환
#    M :  --> MarkDown 셀로 변환
#      1 ~ 5: -> 제목1 ~ 5단계
#    CTRL + S : 저장
```

### Command Mode 단축키
- `m` : Markdown 변환
- `y` : Code 변환
- `A` : 위에 셀 추가
- `B` : 아래에 셀 추가
- `X` : 셀 삭제
- `Z` : 삭제된 셀 복원
- `C` : 셀(들) 복사
- `V` : 복사된 셀(들)은 아래에 붙이기
- `SHIFT + V` -> 복사된 셀(들)을 위에 붙이기
- `1` ~ `5` -> 제목1 ~ 제목5 단계
- `CTRL + S` 혹은 `S` : 저장

### 실행  단축키
- `SHIFT + ENTER` : 실행하고 다음셀로 이동 (없으면 생성)
- `ALT + ENTER` : 실행하고 다음에 새로운 셀 생성
- `CTRL + ENTER` : 해당셀만 실행. 셀이동 안함.

### 셀 병합 / 쪼개기
- `SHIFT + M` : Command mode 에서 선택된 셀들 병합 (merge)
- `SHIFT + CTRL + -` : Edit Mode  에서 커서가 있던 위치 기준으로 셀 쪼개짐 (split)

### EditMode 단축키
- `CTRL + /` : 주석토글
- Line 선택한뒤 `TAB` / `SHIFT + TAB` : 들여쓰기 / 내어쓰기
- 함수, 클래스명 뒤에서 `SHIFT + TAB` : 도움말 docstring
- 변수, 함수, 클래스 작성중 `TAB` : 자동완성
