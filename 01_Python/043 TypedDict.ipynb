# TypedDict

```python
# dict 는 자유롭다.  key-value 쌍

# user 는 name 과  age 키만! 갖게 하고 싶다면? 💢

user = {
    'name': 'Alice',
    'age': 20,    # age 키값은 정수만 허용하게끔 할수 있나?  💢
}
```

```python
user['age'] = '스무살'   # 실행이 되긴 하다..
```

# TypedDict 란
- 딕셔너리의 '구조'와 '타입' 을 미리 정의하는 타입 힌트
- Python 3.8+ 
- typing 모듈에 포함
- 정적 타입 검사 도구(mypy, pyright) 와 함께 사용

```python
from typing import TypedDict
```

```python
class User(TypedDict):
    name: str
    age: int
    
```

```python
# 이제 User 는 정해진 형태의 dict 타입이 된다
```

```python
User(name='John', age=23)
```

```python
User(name='최홍묵)
```

```python
User()
```

```python
type(User())
```

```python
user1 = User(name="박영진", age="스무살")
user1
```
