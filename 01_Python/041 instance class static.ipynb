```python
# 파이썬에서 클래스 내의 변수나 메소드가 선언되는 위치와 방식에 따라 
# 변수는 -> 인스턴스 변수, 클래스 변수 로 나뉩니다. 
# 메소드는 -> 인스턴스 메소드, 클래스 메소드, static(정적) 메소드로 나뉨
# ※ 파이썬은 다른 언어(C++, Java)와 약간 다른 메커니즘을 가지고 있다.

# 🟠 인스턴스 변수 (Instance Variable)
# 각 객체(인스턴스)마다 따로 가지는 변수 -> 인스턴스 네임스페이스에 생성
# - self를 통해 정의, self.변수이름
# - 객체마다 값이 다름 (객체마다 독립적)

# 🟠 클래스 변수 (Class Variable)
# 모든 인스턴스가 공유하는 변수 (공통 데이터 저장용)
# - 클래스 변수는 '클래스에 딱 하나' 생성  --> 클래스 네임스페이스에 생성
# - 클래스 내부에 직접 선언 (self 가 안 붙음)
# - 모든 객체가 같은 값을 참조 (한 곳에서 바꾸면 전체에 영향)
# - ⭐️클래스변수는 가.급.적   '클래스이름.클래스변수이름'으로  사용하자.


# ※ static 변수??
#  파이썬에는 C++/Java처럼 명시적인 static 변수 개념은 없음
#  걍 '클래스 변수'가 그 역할 담당...

```

# 클래스메소드, 인스턴스 메소드, static 메소드
class method,  instance method, static 메소드


```python
# 🟡 인스턴스 메서드 (기본)
#   첫 번째 인자로 self 를 받음 → 인스턴스(객체) 를 가리킴
#   객체의 상태(self.value)를 다루는 용도로 사용

# 🟡 클래스 메서드 (@classmethod)
#  첫 번째 인자로 cls 를 받음 → 클래스 자체를 가리킴
#    클래스 레벨의 데이터(class_var)를 다루거나, 팩토리 메서드 구현 시

# 🟡 정적 메서드 (@staticmethod)
#    self 나 cls 를 받지 않음 → 그냥 클래스 안에 넣은 일반 함수처럼 동작
#    self/cls와 무관한 기능을 클래스 안에 묶고 싶을 때 (ex: 도우미 함수)
```

```python
class MyClass:

    class_var = "I am a class variable"  # 클래스 변수

    def __init__(self, value):
        self.value = value   # 인스턴스 변수

    # 1. 인스턴스 메소드
    # self 가 붙은 메소드 ,  '인스턴스 생성' 하여 사용 
    def instance_method(self):
        return f"[instance_method] self.value = {self.value}"

    # 2. 클래스 메소드
    # 첫 번째 인자로 cls 를 받음 → 클래스 자체를 가리킴
    # self 를 안 받기 때문에 인스턴스 변수 사용 불가.
    @classmethod
    def class_method(cls):
        return f"[class_method] cls.class_var = {cls.class_var}"
    
    # 3. 정적(static) 메서드
    # self 가 안붙은 
    @staticmethod
    def static_method(x, y):
        return f"[static_method] {x} + {y} = {x + y}"    

    def 그냥_method(x, y):
        return f"[그냥_method] {x} * {y} = {x * y}"


obj = MyClass(42)
print(obj.instance_method())
print(obj.class_method()) # 인스턴스로 호출이 되긴하나, 인스턴스 정보는 전달 안됨.
# 클래스 메소드도 클래스.메소드 <- 이와같이 사용하자!
print(MyClass.class_method())

print(obj.static_method(3, 5)) # 정적 메소드 -> self, cls 둘다 안씀 (단순한 독립함수)
print(MyClass.static_method(3, 5))
```

# 과제: PartTimer 설계

```python
# '매 직원(PartTimer)'에 공통적으로 적용되는 자료
# - 시급(공통)
# - 전체 직원수

# 각 직원별 객체 생성시 직원별로 '별칭'과 '근무지' '급여총액' 초기화 (속성)
#   '근무지' 생략시 '113동' 으로 지정
#   직원별로 '급여총액'  0으로 초기화

# 직원의 급여 계산하기(동작)
#    '몇시간 근무',  '+상여금'  에 따른 직원급여 계산
#   '상여금' 은 지정안하면 0 으로 처리


# 예시]
# park = PartTimer('라이언')   // park 은 ‘라이언’ 이라는 닉네임의 직원으로 등록

# park 이 4시간 일한 급여 총액은?  → 34400
# park 이 2시간 일한 급여 총액은? → 17200
# park 이 2시간 일한 급여 + 상여급 2000 총액은? → 19200
```

```python
class PartTimer:
    # TODO
    hourly_wage = 34400 // 4 # 공통 시급
    total_part_timers = 0 # 전체 직원 수

    def __init__(self, nickname, workplace="113동", total_wage=0): # 직원별로 '급여총액'  0으로 초기화
        self.nickname = nickname
        self.workplace = workplace
        self.total_wage = total_wage
        PartTimer.total_part_timers += 1 # 객체 생성시 직원수 +1
        
    def calculate_wage(self, hour, bonus=0):
        r = (PartTimer.hourly_wage * hour) + bonus # 각 인스턴스의 인스턴스변수에 리턴하기위해 변수에 결과 대입
        self.total_wage += r # 총 급여에 +
        return r

    
```

```python
ryan = PartTimer('라이언')
neo = PartTimer('네오', '127-1동')
```

```python
PartTimer.__dict__
```

```python
ryan.total_wage  # 0
```

```python
neo.total_wage  # 0
```

```python
neo.nickname # '라이언'
```

```python
neo.workplace # '127-1동'
```

```python
PartTimer.total_part_timers   # 2    전체직원수
```

```python
ryan.calculate_wage(4)  # 4시간 근무한 급여계산
```

```python
neo.calculate_wage(3, 15000)  # 3시간 근무 + 보너스 급여 계산
```

```python
neo.total_wage  # 0
```
