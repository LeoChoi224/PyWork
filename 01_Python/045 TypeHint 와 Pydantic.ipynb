# 🟥 타입 힌트 (Type Hint)
공식: https://docs.python.org/3/library/typing.html

- Type annotation 이라고도 함


```python
# 파이썬은 동적 타이핑(dynamic typing) 언어다, 
#   변수 타입을 명시적으로 선언하지 않아도 프로그램이 실행될 수 있는 유연성을 가지고 있습니다. 
#   그러나 이러한 유연성은 코드의 가독성과 유지 보수, 에러 발견에 어려움을 초래할 수도 있다.

#   이런 문제를 해결하기 위해 파이썬 3.5부터 도입된 '타입 어노테이션'은 
#   변수와 함수의 반환값, 인자의 타입을 명시적으로 지정할 수 있게 한다.


# [장점]
#   가독성: 타입 정보를 명시하면 코드를 처음 접하는 사람들도 
#         함수의 동작이나 변수의 역할을 쉽게 이해할 수 있습니다.
#   에러 예방: 명시적인 타입 정보를 제공함으로써 개발자들은 
#         미리 잘못된 타입 사용에 대한 문제를 인식하고 수정할 수 있습니다.
#   향상된 도구 지원: 타입 어노테이션을 사용하면 통합 개발 환경(IDE) 등의 도구가 
#         더 정확한 자동 완성, 코드 검사 등을 제공할 수 있습니다.
#   문서화: 타입 정보를 포함하면 주석이나 별도의 문서를 작성하지 않아도 
#         함수의 인터페이스를 이해하기 쉽습니다.

# [단점]
#    추가적인 작업: 타입 정보를 작성하는 데 시간과 노력이 필요. 
#          특히 복잡한 프로젝트의 경우, 타입 정보를 유지하는 데 추가적인 비용이 발생할 수 있다.
# 
#     가독성 저하: 많은 양의 타입 정보가 코드에 포함될 경우, 코드가 지나치게 복잡해 보일 수 있다.


# <<결론>>
#   타입 어노테이션은 프로그래머의 선택에 따라 적용할 수 있는 선택적 기능입니다. 
#   프로젝트의 요구사항과 개발자들의 선호에 따라 타입 어노테이션을 도입할지 여부를 결정할 수 있습니다. 
#   전체적으로 타입 어노테이션은 코드의 가독성, 안정성 및 유지 보수성을 향상시키는데 도움이 됩니다.
```

## class 에서 타입힌트

```python
class BankAccount:
    def __init__(self, account_name: str, initial_balance: int = 0) -> None:
        # instance 변수들의 타입은, type hint 가 있는 매개변수가 대입될때 type hint 가 유추됨.
        self.account_name = account_name
        self.balance = initial_balance
        self.audit_log: list[str] = []

    # self 는 type hint 생략함
    def deposit(self, amount: int) -> None:
        self.balance += amount

    def withdraw(self, amount: int) -> None:
        self.balance -= amount

account: BankAccount = BankAccount("Alice", 400)

def transfer(src: BankAccount, dst: BankAccount, amount: int) -> None:
    src.withdraw(amount)
    dst.deposit(amount)
```

## 필드 설정, 필수, 기본값, Optional, Field

```python
# 일반적으로 Pydantic 모델의 필드는 다음과 같이 정의

# ■ 기본값 제공: field_name: str = "default_value"
#    - 이 경우 필드는 '선택적(optional)', 제공된 기본값을 가집니다.

# ■ 기본값 없음 (필수): field_name: str
#    - 이 경우 필드는 '필수'. 값을 제공하지 않으면 Pydantic은 유효성 검사 오류를 발생시킵니다.

# ■  Optional과 함께 사용: field_name: Optional[str] = None
#   - 이 경우 필드는 '선택적', 명시적으로 None을 기본값으로 가집니다.

# ■ Field와 함께 default 제공: field_name: str = Field(default값)
#    - 이 경우 필드는 '선택적', 제공된 기본값을 가집니다.

# ■ Field와 함께 '...' 사용: field_name: str = Field(...)
#    - 이 경우 필드는 '필수'. 
#    - ... 는 "기본값 없음" 또는 "값이 명시적으로 제공되어야 함"을 나타내는 Pydantic의 특별한 상수로 작동.

```

```python
from pydantic import Field

class User(BaseModel):
    id: int  # 필수 <- None 값 허용 안함
    name: str = '박영진'  # 기본값 설정, optional
    age: Optional[int] = None  # optional
    grade: Optional[int]   # 필수, None 허용
    address: str = Field('hello')  # optional, 기본값, None 허용안함
    nick: str = Field(...) # 필수

User(id=2, name='김정준', age=67, grade=None, nick='GOO')
```

## Custom Validator

```python
# 기본적인 타입 검사 외에도, Pydantic은 각 필드에 대해 사용자 정의 유효성 검사를 추가할 수 있다. 
# 이를 위해 @validator 데코레이터를 사용하여 필드의 값이 특정 조건을 만족하도록 설정할 수 있다.
```

```python
from pydantic import field_validator
```

```python
class Product(BaseModel):
    name: str
    price: float

    # price 필드는 0보다 커야 한다는 검증조건
    @field_validator("price")
    def price_must_be_positive(cls, v):  # class 메소드가 된다. cls 는 클래스!
        print("🐶: price 검증 호출", v)
        if v <= 0:
            raise ValueError("Price must be greater than 0")  # 검증 에러 발생

        return v # 검증 통과


Product(name='맥북', price=-1)
```
