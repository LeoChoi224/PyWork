# 예외 처리 Try ~ except

```python
# 물론 사전에 if 등의 조건식을 활용해서 오류 를 사전에 예방은 가능하나

a = 10
b = 0
if b != 0:
    print(a / b)
else:
    print('0 으로 나눌수 없습니다')

# 수~~~많은 예외상황을 매번 if 조건문으로 처리하면
# 코드가 매우 난잡해진다. --> 유지보수 힘들어짐.

# 그래서, 예외처리 관련하여
# '코드수행블럭'  와 '예외처리블럭' 구분하여 예외를 다루는게 바람직.
```

```python
# 파이썬 에서는 
# try ~ except 문으로 오류를 다룹니다

# 구문]
# try:
#     코드수행블럭
#     ...
# except [발생 오류[as 오류 메시지 변수]]:
#     예외처리블럭
#     ...

#  try 블록 수행 중 오류가 발생하면 except 블록이 수행된다. 
# 하지만 try블록에서 오류가 발생하지 않는다면 except 블록은 수행되지 않는다.

```

## try ~ except¶

```python
num = 0

try:
    # 코드 수행블럭. 이 안에서 예외 발생하면 except: 으로 넘어감.
    print('1. try 시작')
    result = 4 / num # 오류 발생하면 try 블럭 나머지는 실행안함
    print('결과', result)
    print('2. try 종료')
except:
    # 예외가 발생하면 실행
    print('3. except 시작')

print('4. 프로그램 정상 종료')
    
```

## try ~ except 오류:

```python
num = 0

try:
    print('1. try 시작')
    result = "4" / num
    print('결과', result)
    print('2. try 종료')
except ZeroDivisionError: # try 블럭에서 ZeroDivisionError 발생하면 여기서 처리
    print('3-1. ZeroDivisionError 처리')
except NameError: # except: 는 여러 개 붙힐 수 있다.
    print('3-2. NameError 처리')
except TypeError:
    print('3-3. TypeError 처리')
print('4. 프로그램 정상 종료')
    
```

## 3. try ~ except 오류 as 변수:

```python
# 예외 객체 안에서 발생죈 예외관련 정보들이 담겨 있다
```

```python
num = 0

try:
    print('1. try 시작')
    result = "4" / num
    print('2. try 종료. 결과', result)
except ZeroDivisionError as e:
    print('3-1. ZeroDivisionError 처리')
    print('🔶', e.args)  # tuple 안에 에러 관련 정보들 담김
    print('🔷', e)
except NameError as e:
    print('3-2. NameError 처리')
    print('🔶', e.args)  # ("name 'n' is not defined",)
    print('🔷', e)
except TypeError as e:
    print("3-3. TypeError 발생")    
    print('🔶', e.args)  # ("unsupported operand type(s) for /: 'str' and 'int'",)
    print('🔷', e)

print('4. 프로그램 정상 종료')   
```

## 4. try ~ finally:
예외가 발생했든 아니듯 반드시 수행해야 하는 코드들은 finally 에 작성

```python
try:
    print('1. try 시작')
    f = open('foo.txt', 'w')
    a = 10
    a = a + bbbbbbb
    print('2. try 종료')
except:
    print('3. 예외처리')
    # 설사 return 을 해도, 그 전에 finally 가 실행된다.
finally:
    # 반드시 수행되는 블럭
    print('4. finally')
    f.close()

print('5. 프로그램 종료')
```

## 5. 여러 오류 한방에 처리하기

```python
# try:
#     ...
# except 발생 오류1:
#    ... 
# except 발생 오류2:
#    ...

# 방법2
# try:
#     ...
# except (발생 오류1, 발생오류 2 ..):
#    ... 
```

```python
num = 0

try:
    print('1. try 시작')
    result = "4" / num
    print('2. try 종료. 결과', result)
except (ZeroDivisionError, NameError) as e:
    print('3-1. ZeroDivisionError, NameError 처리', e.__class__.__name__)
    print('🔶', e.args)
    print('🔷', e)
except Exception as e:  # 나머지 발생되는 모든 에러는 여기서 처리
    print("3-3. 기타 에러들 발생")    
    print('🔶', e.args)
    print('🔷', e)

print('4. 프로그램 정상 종료') 
```

# 강제로 오류 발생시키기 raise

```python
# 파이썬 에서는 
# try ~ except 문으로 오류를 다룹니다

# 구문]
# try:
#     코드수행블럭
#     ...
# except [발생 오류[as 오류 메시지 변수]]:
#     예외처리블럭
#     ...

#  try 블록 수행 중 오류가 발생하면 except 블록이 수행된다. 
# 하지만 try블록에서 오류가 발생하지 않는다면 except 블록은 수행되지 않는다.
```

```python
#  Bird라는 클래스를 상속받는 자식 클래스는 
# 반드시 fly라는 함수를 구현하도록 만들고 싶은 경우(강제로 그렇게 하고 싶은 경우)가 있을 수 있다. 다음 예를 보자
# '구현(implement)' : 부모클래스의 메소드를 오버라이딩해서 재정의하는것.

class Bird:
    def fly(self):
        raise NotImplementedError
        
```

```python
class Eagle(Bird):
    pass

eagle = Eagle()
eagle.fly()
print('🟨 프로그램 종료')
```

```python
class Eagle(Bird):
    # 상속 받은 클래스에서 fly 구현 (오버라이딩)
    def fly(self):
        print('very fast')

eagle = Eagle()
eagle.fly()
print('🟨 프로그램 종료')
```

# 예외의 전파 (Propagation)

```python
# 예외가 처리(handling) 되는 순서

# 1. 예외 발생 → 현재 코드 블록 확인2
# 2. 처리 안 되면 호출한 함수로 전파
# 3. 계속 위로 올라감 (호출 스택 따라감)
# 4. 어디서도 처리하지 않으면 프로그램 종료 + traceback 출력
```

```python
def f():
    raise ValueError("문제 발생!")  # 예외 발생. 발생된 블럭에서 처리 안함

f()  # 호출한 쪽으로 예외 전파.  여기서도 처리 안됨 -> 프로그램 종료 + traceback 출력

print('🟨 프로그램 종료')
```

```python
def f():
    raise ValueError("문제 발생!") 

def g():
    f()

def h():
    g()
# -------------------------------
h()
print('\n🟨 프로그램 종료')
```

```python
def f():
    raise ValueError("문제 발생!", "누가 문제일까?", "태현이가?", "민재가?", "홍목이가?") 

def g():
    f()

def h():
    g()
# -------------------------------
try:
    h()
except ValueError as e:
    print(e.args)

print('\n🟨 프로그램 종료')
```

# 커스텀 오류 만들기

```python
# Exception 을 상속하여 만들 수 있다.
```

```python
class MyError(Exception):
    def __init__(self, msg):
        self.msg = msg

    def __str__(self):   # 에러 메세지
        return self.msg
```

```python
def say_nick(nick):
    if nick in ['바보', '멍청이']:
        raise MyError(f"MyError: {nick} 은 허용하지 않는 별명")

    print(f'say_nick({nick})')

try:
    say_nick('천사')
    say_nick('바보')
except MyError as e:
    print(e)  # __str__

print('프로그램 종료')

    
```

```python
# --------------------------------------------------------------------------
# 점수를 입력받기
# 유효한 점수는 0 ~ 100점
# 입력점수가 유효하지 않으면 ScoreException 발생시키기

class ScoreException(Exception):
    def __init__(self, msg="점수입력오류"):
        self.msg = msg

    def __str__(self):   # 에러 메세지
        return self.msg    


def input_score():
    score = int(input('점수를 입력하세요(0~100):'))

    if score < 0 or 100 < score:
        raise ScoreException(f'{score} 값은 유효하지 않은 점수입니다 (0~100)')

    return score

try:
    print('국어, 영어, 수학 점수를 차례로 입력하세요')
    kor = input_score()
    eng = input_score()
    math = input_score()
    print(f'총점: {kor + eng + math}')
    print(f'평균: {(kor + eng + math) / 3}')
except ScoreException as e:
    print(e)

print('프로그램 종료')
```

# 실습: 나이 입력 받기 AgeException

```python
# 나이 입력받기
# 유효한 나이는 0이상의 값
# 유효하지 않은 나이를 입력받으면 AgeException 발생
# 정상적인 나이를 입력받을때까지 예외 처리후 계속 입력하게 하려면?

"""
나이를 입력하세요(0 <):-4
-4 는 유효하지 않은 나이입니다   <- 예외 메세지
다시 입력하세요                 <- 예외 발생하면 재입력 유도
나이를 입력하세요(0 <):34
입력하신 나이는 34 입니다
프로그램 종료
"""
None
```

```python
class AgeException(Exception):
    def __init__(self, msg="나이입력오류"):
        self.msg = msg

    def __str__(self):
        return self.msg

def input_age():
    age = int(input("나이를 입력하세요(0 <): "))

    if age < 0:
        raise AgeException(f"{age} 는 유효하지 않은 나이입니다")

    return age

while True:   
    try:
        age = input_age()
        print(f"입력하신 나이는 {age} 입니다")
        break
    except AgeException as e:
        print(e)
        print("다시 입력하세요")
    
print("프로그램 종료")       
    
```
