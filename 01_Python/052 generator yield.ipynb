```python
def number_function():
    return 0

number_function()
```

```python
def number_generator():
    yield 0    # 함수 안에서 yield 를 사용하면 이 함수는 genearator 가 된다.

number_generator()
```

```python
def number_generator():
    print('🐹 gen1')
    yield 0
    print('🐹 gen2')
    yield 1
    print('🐹 gen3')
    yield 2
    print('🐹 gen4')

for i in number_generator():   # generator 함수는 iterator 를 생성해줌.
    print('🍊 for i=', i)
```

```python
g = number_generator()
dir(g)
```

```python
# generator는 iterator 와 동작은 같다..

# [차이점]
# iterator 에선 __next__() 안에서 직접 return 값으로 반환했었다
# generator 에선 yield 로 지정한 값이 __next__() 의 리턴값으로 동작한다.

# iterator 에선 raise StopIteration 으로 예외를 직접 발생
# generator 에선 함수 끝까지 도달하면 StopIteration 이 자동으로 발생
```

왜 **yield** 라고 이름 지었을까?  양보하다란 뜻<br>
즉 **yield 값** 으로 함수 바깥으로 값을 전달하면서 코드 실행을 함수 바깥에 양보,<br>
현재 함수는 잠시 중단하고 함수 바깥의 코드가 실행됨.


```python
def enum_alpha(n):
    txt = "abcdefghijklmnopqrstuvwxyz"
    for ch in txt[:n]:
        yield ch

for ch in enum_alpha(4):     # a, b, c, d
    print(ch, end = ' ')
```

### generator 와 return
generator 는 함수 끝나면 StopIteration 예외 발생<br>
return 도 함수를 끝내므로 중간에 return 만나면 StopIteration 발생<br>
`return 값` : StopIteration 의 예외메세지로 동작함



```python
def number_generator():
    yield 0
    yield 1
    return '📌generator 종료'
    yield 2


try:
    g = number_generator()
    while True:
        print('🎃', next(g))
except StopIteration as e:
    print('StopIteration', e)

```

## 실습: range(n) 과 같은 동작을 하는 generator 만들기

```python
# 도전: generator 응용
# range(n) 와 같은 동작을 하는 generator 만들기

def number_generator(stop):
    n = 0
    while n < stop:
        yield n
        n += 1

for i in number_generator(5):
    print(i)
    
# 결과
# 0
# 1
# 2
```

## yield from 

```python
# yield from 사용 (파이썬 3.3 <= )
# 여러번 바깥으로 전달하기

#   yield from iterable
#   yield from iterator
#   yield from generator
```

```python
# 지금까지는 yield 를 사용하여 값을 한번에 '하나씩' 바깥으로 전달햇었다
def number_generator():
    x = [1, 2, 3]
    for i in x:
        yield i

for i in number_generator():
    print(i)
```

```python
def number_generator():
    x = [1, 2, 3]
    yield from x

for i in number_generator():
    print(i)
```

# Generator Expression
list comprehension 은 **[  ]** 를 사용했지만<br>
같은 표현식을 **(  )** 로 묶으면 **generator expression** 이 됩니다.

list comprehension 은 list 를 만들어내지만 --> 메모리에 생성됨(공간 차지)👿<br>
**generator expression** 은 필요할때만 요소를 만들어낸다 --> 메모리 절약 😀


```python
d = [i for i in range(0, 100, 2)]
sum(d)  # sum(iterable)
```

```python
sum(i for i in range(0, 100, 2))  # generator expression
```

```python
n = 500

d = [i for i in range(n) if i % 2 == 0]
print(d)
g = (i for i in range(n) if i % 2 == 0)
print(list(g))

import sys
print(sys.getsizeof(d)) # byte 용량
print(sys.getsizeof(g))
```
