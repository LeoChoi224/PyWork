# 🟦os 모듈

```python
import os
```

```python
os.path
```

```python
savedir = 'tmpdir1'
```

```python
# exists()  디렉토리 존재 여부
# makedirs() 는 한꺼번에 중간경로(intermediate directorieds) 들 생성.
# mkdir() 은 단일 디렉토리만 생성

if not os.path.exists(savedir):
    os.makedirs(savedir)
    print(savedir, '폴더 생성')
else:
    print(savedir, '폴더는 이미 존재합니다.')
```

```python
# 임시 파일 생성
filelist = [
    os.path.join(savedir, 'file1.txt'),
    os.path.join(savedir, 'file2.txt'),
]

for file in filelist :
    with open(file, 'w') as f:
        pass
```

```python
!ls -ll tmpdir1/*
```

```python
os.path.dirname(filelist[0])  # 상대경로
```

```python
os.path.abspath(filelist[0])  # 절대경로.
```

```python
filename = os.path.basename(filelist[0])  # 경로에서 파일명 (가장 끝의 경로.)
filename
```

```python
os.path.splitext(filename)  # 파일명과 확장자 구분. -> Tuple
```

```python
os.listdir(savedir)  # 주어진 경로상의 파일/디렉토리 목록 -> List
```

```python
os.listdir()  # 매개변수 없으면 현재 작업경로 이하의 목록
```

```python
for d in os.listdir():
    if os.path.isdir(d):
        print(d, "🟣directory")
    if os.path.isfile(d):
        print(d, "💚file")
```

```python
filelist
```

```python
for file in filelist:
    os.remove(file)
    print(file, '을 삭제했습니다')
```

```python
!ls -ll tmpdir1/*
```

```python
os.removedirs(savedir)  # 디렉토리 삭제
```

```python
os.sep # 경로 구분 문자.
```

# 🟦 shutil 모듈

```python
import shutil
```

```python
# 실습용 폴더 만들기
test_dir3 = "tmpdir3"
if not os.path.exists(test_dir3):
    os.makedirs(test_dir3)
```

```python
os.getcwd()
```

```python
os.chdir(test_dir3) # 현재 작업경로 변경
os.getcwd()
```

```python
# 실습용 폴더에 실습용 파일 만들기
with open("sample.txt", "w") as f:
    for i in range(1, 11):
        data = "Line %d\n" % i
        f.write(data)
```

```python
!ls -all
```

```python
shutil.copy('sample.txt', 'sample_copy.txt')
```

```python
!ls -all
```
