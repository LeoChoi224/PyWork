```python
import requests
```

```python
key = "484e536751686d6336306d484d7576"
date = "20260401"
start_index = 1
end_index = 10
```

# json

```python
req_type = 'json'

url = f'http://openapi.seoul.go.kr:8088/{key}/{req_type}/CardSubwayStatsNew/{start_index}/{end_index}/{date}'

url
```

```python
response = requests.get(url)  # url 로 요청 -> 응답 response 리턴
response # Response 객체
```

```python
# <Response [200]>  <- 응답코드값 200: 정상처리 (200이라 해서 에러가 없다는 뜻은 아니다)
```

```python
response.status_code# JSON 텍스트 -> 파이썬 데이터 변환
data = response.json()
data
```

```python
rows = data['CardSubwayStatsNew']['row']
len(rows)
```
