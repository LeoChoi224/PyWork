# JSON

```python
"""
  JSON : JavaScript Object Notation
 	데이터 교환 용도의 JavaScript object 구문
 	오늘날 웹, 앱 등 애플리케이션 개발에서 데이터교환의 중요한 ★텍스트★포맷.


※파이썬과 자바스크립트의 데이터구문은 거의 모양은 동일하나, 
명칭이 조금 다른다.
 
JSON              파이썬  |  자바스크립트
──────────────────────────────────────
숫자           int, float  |    number
문자열            str      |    string
[]               list     |    배열(array)
{name:value..}    dict     |    object
true / false      bool     |    boolean
null              None     |    null


 	JSON 구문에선  
 		name/value 쌍에서 name은 반드시 쌍따옴표로 감싸야 한다
 		문자열 데이터도 반드시 쌍따옴표로만 감싸야 한다.
 			ex) {"name":"John"}

        배열이나 object 의 마지막 원소 뒤에 콤마 허용 안함.
 		 			
 	JSON 으로 담을수 있는 파이썬 data type 들
 		str, int, flost, list, tuple, dict, None
 	
 	JSON 으로 담을수 없는 data type 들
 		set, function, datetime


파이썬에선 JSON 을 다루기 위한 'json' 모듈을 기본 제공한다
"""
```

```python
import json
```

# dumps, loads

```python
# 파이썬 객체 -> JSON 문자열
json.dumps('kim')
```

```python
json.dumps(100)
```

```python
json.dumps('김정준')  # 한글은 인코딩되어 변환.  유니코드 이스케이프 시퀀스 표현. 
                      # UTF-16 코드 유닛을 ASCII 범위내 문자로 표현.
```

```python
json.dumps('김정준', ensure_ascii=False)  # 한글이 인코딩 되지 않고 저장됨.
```

```python
for d in [10, 3.14, "Hello", '철수',
          True,False,
          None,
          [10, 20, 30,], ('a', 'b', 'c'),
          {'name': 'John', 'age': 14},
         ]:
    
    print(f"{str(d):>30} → {json.dumps(d, ensure_ascii = False)}")  #  ensure_ascii = False  <- 한글이 인코딩 안되어 저장되게 하기
```

```python
json.loads("[4, 5, 6]")  # JSON 텍스트 -> 파이썬 객체
```

# dump, load
파이썬 데이터 ↔ json파일

```python
import os
base_path = r'out'
file_path = os.path.join(base_path, 'jsondata.json')  # *.json 확장자
```

```python
data = {
    "name": "홍길동",
    "birth": "2025-03-21",
    "age": 30,
}

```

```python
with open(file_path, 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False)
```

```python
!type out/jsondata.json
```

```python
with open(file_path, encoding='utf-8') as f:
    data = json.load(f)

data
```
