# 파일 (File) 다루기

```python
# 파일을 다루기 위해선 open(), close() 함수로 감싸여진다

# 파일객체 = open(파일 이름, 파일 열기 모드)
#   ... 파일객체사용 (읽기 or 쓰기)
#   ... 파일객체사용 (읽기 or 쓰기)
#   ... 파일객체사용 (읽기 or 쓰기)
# 파일객체.close()

# 시스템 자원(resource) 에 대한 프로그래밍은 일반적으로 open ~ close 형태로 작업.
#   ex) 파일, 네트워크, 데이터베이스, ...
```

```python
# 파일 열기 모드
# r  읽기모드 - 파일을 읽기만 할 때 사용
# w  쓰기모드 - 파일에 내용을 쓸 때 사용.  해당 파일이 없으면 새로 생성.   해당 파일이 있었으면 삭제하고 새로 생성 ★
# a  추가모드 - 파일의 마지막에 새로운 내용을 추가 시킬 때 사용.  해당 파일이 없으면 새로 생성
```

## 쓰기

```python
f = open('새파일.txt', 'w')

for i in range(1, 11):
    data = f"Line {i}\n"
    f.write(data)  # 파일에 쓰기
    
f.close()
```

## 읽기

```python
f = open('새파일.txt', 'r')   # 읽기모드 'r'

line = f.readline()  # 한 줄 읽기
print(line)

f.close()
```

```python
f = open('새파일.txt', 'r')   # 읽기모드 'r'

while True:
    line = f.readline()  # readline() 은 더 이상 읽을 라인이 없으면 None 리턴
    if not line: break   # 파일 끝까지 읽었으면 종료
    print(line, end="")
f.close()
```

```python
f = open('새파일.txt', 'r')   # 읽기모드 'r'

lines = f.readlines()   # 각 line 을 담은 list 리턴
print(''.join(lines))

f.close()
```

```python
f = open('새파일.txt', 'r')   # 읽기모드 'r'

# file 객체도 iterable 하다. 한 line 씩 나옴

for line in f:
    print(line,end='')

f.close()
```

```python
f = open('새파일.txt', 'r')   # 읽기모드 'r'

data = f.read() # 파일 전체 읽어옴.
print(data)

f.close()
```

```python
# Windows 는 시스템 인코딩이 기본 2byte (cp949, ms949, ANSI, euc-kr)이다
# 특별히 encoding 옵션을  지정하지 않으면 파일은 2byte 로 인코딩되어 있다
# ※ 반면 MAC, Linux 등에선 기본인코딩이 3byte (utf8) 이다
```

```python
with open(r'out/새파일.txt', 'w') as f:
    for i in range(1, 11):
        data = f"{i}번째 줄입니다\n"
        f.write(data)
```

```python
ll out/*.txt
```

```python
with open(r'out/새파일.txt', 'r') as f:
    print(f.read())
```

```python
# 명시적으로 encoding 옵션지정
with open(r'out/새파일utf-8.txt', 'w', encoding='utf8', errors='ignore') as f:
    for i in range(1, 11):
        data = f"{i}번째 줄입니다\n"
        f.write(data)
```

```python
with open(r'out/새파일utf-8.txt', 'r', errors='ignore') as f:
    print(f.read())
```

```python
# result = {
#    1: [0, 1, 1, 0],
#    2: [1, 1, 1, 1, 1, 0, 0],
#    3: [0, 1],
# }

result = {}

with open(file_path, "r", encoding="utf-8") as f:
    
    f.readline()  # 첫 라인 (컬럼헤더) 는 skip
    for line in f: # line 별로 읽어오기  "1,0,3,"Braund, Mr. Owen Harris",male,22,1,0,A/5 21171,7.25,,S"
        # "," 콤마로 쪼개기
        psgr = line.split(',')  # ['1', '0', '3', '"Braund', ' Mr. Owen Harris"', 'male', '22', '1', '0', 'A/5 21171', '7.25', '', 'S\n']
        
        # 2번째 컬럼 -> survived
        # 3번째 컬럼 -> pclass
        survived = int(psgr[1])    # "1", "0"
        pclass = int(psgr[2])      # "1", "2", "3"
        if result.get(pclass):
            result[pclass].append(survived)
        else:
            result[pclass] = [survived]

        # print(f'{pclass}-{survived} => {result}')
        # input()

for pclass in sorted(result):
    total = len(result[pclass])  # pclass 등급 객실의 전체 탑승객수
    survived = result[pclass].count(1)  # pclass 등급 탑승객중 생존한 분들의 숫자
    print("{}등급] 총 {} 명, 생존 {} 명, 생존률 {:.1f}%".format(pclass, total, survived, (survived / total) * 100 ))

```
